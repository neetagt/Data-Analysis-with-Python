Numerical Analysis and Visualization

```python
sales['Unit_Cost'].describe()
```

```python
sales['Unit_Cost'].mean()
```

```python
sales['Unit_Cost'].median()
```

```python
sales['Unit_Cost'].plot(kind='box',vert = False,figsize=(12,5))
```

```python
sales['Unit_Cost'].plot(kind='density',figsize=(12,5))
```

```python
ax = sales['Unit_Cost'].plot(kind='density',figsize=(12,5))
ax.axvline(sales['Unit_Cost'].mean(),color='red')
ax.axvline(sales['Unit_Cost'].median(),color='green')
```

```python
ax = sales['Unit_Cost'].plot(kind='hist',figsize=(12,5))
ax.set_xlabel('dollars')
ax.set_ylabel('Number of Sales')
```

```python
subset_corr = sales[['Date', 'Day', 'Year','Customer_Age','Order_Quantity','Unit_Cost','Unit_Price','Profit','Cost','Revenue']].corr()
print(subset_corr)
```

```python
fig = plt.figure(figsize=(8,8))
plt.matshow(subset_corr,cmap='RdBu')
plt.xticks(range(len(subset_corr.columns)), subset_corr.columns,rotation = 'vertical')
plt.yticks(range(len(subset_corr.columns)), subset_corr.columns)
```

plt.figure():
This function creates a new "Figure" object, which serves as the top-level container for all plot elements.
plt.plot():
This function is used to draw 2D plots within an "Axes" object. It takes data (e.g., x and y coordinates) and plots them as lines, markers, or a combination of both.

```python
sales.plot(kind='scatter',x='Customer_Age',y='Revenue',figsize=(8,4))
```

```python
sales.plot(kind='scatter',x='Revenue',y='Profit',figsize=(6,6))
```

```python
ax = sales[['Profit','Age_Group']].boxplot(by='Age_Group',figsize=(10,6))
ax.set_ylabel('Profit')
```

```python
boxplot_cols = ['Year','Customer_Age','Order_Quantity','Unit_Cost','Unit_Price','Profit']
sales[boxplot_cols].plot(kind = 'box',subplots = True,layout=(2,3),figsize=(14,8))
```
