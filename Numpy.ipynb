Differences between Numpy and List:

***Numpy is faster thank List because Numpy has fixed type(binary) whereas List has object type. Hence faster to read because less bytes of memory and no type-checking while iterating through objects.

***Contiguous memory is used in Numpy which helps in SIMD(Single Instruction Multiple Data) Vector Processing i.e. we can perform operations on all the bits(units) at the same time. It also helps in effective cache utilization.

***More applications of numpy e.g., Mathematics, Plotting, Backend(Pandas, Connect 4, Digital Photography), Machine Learning,

###Copying arrays

```python
a = np.array([1,2,3])
b = a
b[0] = 100
print(b)
print(a)
```

```python
a = np.array([1,2,3])
b = a.copy()
b[0]=100
print(b)
print(a)
```
