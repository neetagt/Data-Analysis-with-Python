By default:
Mean () = 0

Standard deviation () = 1

size=2 means you want 2 samples.

np.empty((2, 2)) creates a 22 NumPy array without initializing its values.

It allocates memory for the array but does not set the elements to any specific value.

The contents will be whatever happens to already exist at that memory location  typically random-looking garbage values.

```python
np.empty(5)
```
