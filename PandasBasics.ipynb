The .quantile() function in Pandas is used to compute the quantile(s) of a Series or DataFrame. A quantile is a value below which a given percentage of observations in a group of observations falls. This is especially useful for statistical analysis.

1. For a Series
import pandas as pd

s = pd.Series([1, 2, 3, 4, 5])
s.quantile(0.5)
Output:
3.0
Explanation: This returns the median (50th percentile) of the series.

2. For a DataFrame
python
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1]
})

df.quantile(0.25)

Output:

A    2.0
B    2.0
Name: 0.25, dtype: float64
Explanation: This returns the 25th percentile for each column.
